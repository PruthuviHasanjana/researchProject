```python
#import methylation dataset
methylation = pd.read_csv("Methylation27.csv",  header=None)

#remove index
# methylation.reset_index('sample',inplace=True)

#transpose the dataset
methylation = methylation.set_index(0).transpose()
```

```python
#Select cancer stage as a target
stage =phenotype[['sampleID','AJCC_Stage_nature2012']] 

#rename primary column
stage = stage.rename(columns={'sampleID': 'sample'})
```

```python
methylationStage = methylation.merge(stage, how ='inner', on='sample')
```

```python
methylationStage
```

```python
methylationStage=methylationStage.dropna(subset=['AJCC_Stage_nature2012']) 
```

```python
methylationStage=methylationStage.fillna(methylationStage.mean())
```
